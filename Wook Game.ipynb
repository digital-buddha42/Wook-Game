```python
BANDS = {
    'ph': 'Phish',
    'spaf': 'Spafford',
    'db': 'The Disco Biscuits',
    'wsp': 'Widespread Panic',
    'gd': 'The Grateful Dead',
    'goo': 'Goose',
}

SCORES = {
    'ph': 3,
    'spaf': 2,
    'db': 3,
    'wsp': 3,
    'gd': 4,
    'goo': 1,
}

def ask_question(question, options):
    print(question)
    for key, value in options.items():
        print(f"{key}) {value}")
    while True:
        answer = input("Enter your answer: ")
        if answer in options:
            return answer
        print("Please enter a valid option.")

def main():
    print("Welcome to the Wook Game!\n")
    print("In this game, I ask you a series of multiple choice questions about your jam-band related tastes.")
    print("After you answer, I tell you if you are heady or just a crusty wook!\n")
    name = input("What is your name? ")

    while True:
        try:
            age = int(input("What is your age in years? "))
        except ValueError:
            print("Please enter a valid integer for age!")
            continue

        if age < 18:
            print("Better get your hippie mom to explain that Dead and Co ain't a John Mayer band!")
        else:
            print(f"\nSup {name}! Hope you brought some granola spunion! Let's jam!\n")
            break

    score = 0

    answer_1 = ask_question("Who is your favorite band?", BANDS)
    score += SCORES[answer_1]
    print(f"{BANDS[answer_1]} is a great choice!\n")
    
    print(f'So far you have {score} points!\n')

if __name__ == "__main__":
    main()

```
